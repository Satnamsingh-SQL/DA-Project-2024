**This DataFrame represents a dataset of music-related information. Here's a brief description of the columns:** 


1. **``Track URI``**: A unique identifier for each track on Spotify.
2. **``Track Name``**: The name/title of the track.
3. **``Artist URI(s)``**: Unique identifier(s) for the artist(s) of the track.
4. **``Artist Name(s)``**: The name(s) of the artist(s) associated with the track.
5. **``Album URI``**: Unique identifier for the album containing the track.
6. **``Album Name``**: The name of the album containing the track. 
7. **``Album Artist URI(s)``**: Unique identifier(s) for the artist(s) of the album.
8. **``Album Artist Name(s)``**: The name(s) of the artist(s) of the album.
9. **``Album Release Date``**: The release date of the album. 10. Album Image URL: URL(s) to the album cover image.
11. **``Disc Number``**: The disc number of the track if part of a multi-disc release.
12. **``Track Number``**: The track number within the album.
13. **``Track Duration (ms)``**: The duration of the track in milliseconds.
14. **``Track Preview URL``**: URL to a preview or sample of the track.
15. **``Explicit``**: A boolean indicating whether the track contains explicit content.
16. **``Popularity``**: A measure of the track's popularity.
17. **``ISRC``**: International Standard Recording Code for the track.
18. **``Added By``**: Information about who added the track (possibly a user or source).
19. **``Added At``**: Date and time when the track was added.
20. **``Artist Genres``**: Genres associated with the artist(s). 
21. **``Danceability``**: A measure of how suitable the track is for dancing.
22. **``Energy``**: A measure of the energy of the track.
23. **``Key``**: The key of the track(major vs minor music keys).
24. **``Loudness``**: A measure of the loudness of the track.
25. **``Mode``**: The modality of the track (major or minor).
26. **``Speechiness``**: A measure of the presence of spoken words in the track.
27. **``Acousticness``**: A measure of the track's acoustic characteristics.
28. **``Instrumentalness``**: A measure of the track's instrumental content.
29. **``Liveness``**: A measure of the presence of a live audience in the track.
30. **``Valence``**: A measure of the track's positivity.
31. **``Tempo``**: The tempo of the track.
32. **``Time Signature``**: The time signature of the track. 
33. **``Album Genres``**: Genres associated with the album (though it appears to be mostly empty in this dataset).
34. **``Label``**: The record label associated with the track.
35. **``Copyrights``**: Copyright information associated with the track. This dataset appears to contain a wide range of information related to tracks, albums, and artists, including audio features that can be used for music analysis and recommendation systems.

```python
# # Convert the 'Album Release Date' column to datetime
# Filter out rows with invalid dates in 'Album Release Date'
df['Album Release Date'] = pd.to_datetime(df['Album Release Date'], errors='coerce')

# Drop rows with invalid dates (NaT) from the DataFrame
df.dropna(subset=['Album Release Date'], inplace=True)


# # Extract the year from the release date
df['Release Year'] = df['Album Release Date'].dt.year

# # Selecting relevant columns for the plot: Release Year and Popularity
data_for_boxplot = df[['Release Year', 'Popularity']]

# # Handle missing values in the 'Artist Genres' column
data_for_boxplot['Release Year'].fillna('Unknown', inplace=True)

# # Create the boxplot
plt.figure(figsize=(12, 5))

sns.boxplot(data=data_for_boxplot, x='Release Year', y='Popularity',hue='Release Year')
plt.xticks(rotation=90)
plt.yticks(range(-50,200,50))
plt.title('Boxplot of Popularity by Release Year')
plt.tight_layout()


from datetime import datetime
date = datetime.now()
date = str(date).replace('-','_').replace(':','_').replace('.','_')

plt.savefig(f'all_findings_images/Boxplot of Popularity by Release Year {date}.png',dpi = 1000)
plt.show()
```
